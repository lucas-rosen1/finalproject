```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
    
mean = pd.read_csv('data/final_means.csv',header=None,index_col=0)
swing_states = mean.loc[['Florida','Michigan','Pennsylvania','Wisconsin']]
starts = ((swing_states-.5)*200).round(2)[1]
```

```python
#DAG MH
swings = [starts]
ecs = []
times = ['Day','7 PM','7:30 PM','8 PM','8:30 PM','9 PM','10 PM','Final']
for st in ['6.5','7.0','7.5','8.0','8.5','9.0','10.0','13.0']:
    with open('data/dag_MH_{}.pkl'.format(st), 'rb') as f:
        swing, ec = pickle.load(f)
        ecs.append(ec)
        if st=='6.5':
            continue
        swings.append(((swing-.5)*200).round(2))
        
        

plt.plot(times,ecs,'x', label = 'Clinton EV Prediction')
plt.plot(times,np.ones(8)*270,label='Win Threshold')
plt.xlabel('State Returns')
plt.ylabel('EVs')
plt.title('Clinton Predicted EVs Using Metropolis-Hastings Sampling and DAG Model')
plt.legend()
plt.show()

frame = pd.DataFrame(swings).T
frame.columns = times
frame.index.name='State'
frame
```
