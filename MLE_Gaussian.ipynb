```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pickle
```

```python
def MLE_dag(dag_mat, data_list, columns):
    """Calculate MLE for DAG and MICE data tables"""
    
    mles_list = []
    for i in range(dag_mat.shape[1]):
        col = dag_mat[:,i]
        evidence = list(columns[col.astype(bool)])
        responce = columns[i]
        
        var_list = []
        beta_list = []
        for data in data_list:
        
            y = data.iloc[:,i]
            X = data.iloc[:,col.astype(bool)]
            reg = LinearRegression()
            pars = X.shape[1]

            ## If no parents, predict the mean.
            if pars==0:
                y_pred = np.zeros(len(y))
                y_pred[:] = y.mean()
                beta_list = []

            else:
                reg.fit(X,y)
                y_pred = reg.predict(X)
                beta_list.append(reg.coef_)


            var_list.append(mean_squared_error(y,y_pred))
        beta = np.array(beta_list).mean(0)
        var = np.mean(var_list)
        mles_list.append((responce, beta, var, evidence))
    return mles_list
```

```python
with open('data/data_fill.pkl','rb') as f:
    data_fill_list = pickle.load(f)
```

```python
pc_dag = pd.read_csv('data/DAG_PC.csv',index_col='Unnamed: 0')
MLE_PC = MLE_dag(pc_dag.values, data_fill_list, pc_dag.columns)
```

```python
## Pickle the list
with open('data/MLE_PC.pkl', 'wb') as f:  
    pickle.dump(MLE_PC, f)
```
