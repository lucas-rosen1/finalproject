```python
from fancyimpute import MICE
import numpy as np
import pandas as pd
import pickle
```

```python
## Load data
data_full = pd.read_csv('data/full_data.csv',index_col='startdate')
```

```python
## Do 10 MICE iters
# Subtract mean to make gaussian easier
data_fill_list = []
means_list = []
for i in range(10):
    data_fill = pd.DataFrame(MICE().complete(
        data_full.values),columns = data_full.columns, 
                        index = data_full.index)
    
    mean = data_fill.mean(0)
    data_fill_list.append(data_fill - mean)
    means_list.append(mean)
    
    
```

```python
## Pickle the list
with open('data/data_fill.pkl', 'wb') as f:  
    pickle.dump(data_fill_list, f)
    
with open('data/means.pkl', 'wb') as f:  
    pickle.dump(means_list, f)
        
```
